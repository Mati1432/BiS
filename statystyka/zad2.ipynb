### Zadanie 2

```python
k = 10
p = 0.5
n = 100
lambda_param = 10

samples_bernoulli = np.random.binomial(1, p, n)

samples_binomial = np.random.binomial(k, p, n)

samples_poisson = np.random.poisson(lambda_param, n)
```

### Zadanie 3

```python
mean_bernoulli = np.mean(samples_bernoulli)
mean_binomial = np.mean(samples_binomial)
mean_poisson = np.mean(samples_poisson)

print(f'{mean_bernoulli}')
print(f'{mean_binomial}')
print(f'{mean_poisson}')
```

```python
var_bernoulli = np.var(samples_bernoulli)
var_binomial = np.var(samples_binomial)
var_poisson = np.var(samples_poisson)

print(f'{var_bernoulli}')
print(f'{var_binomial}')
print(f'{var_poisson}')
```

```python


kurt_bernoulli = scipy.stats.kurtosis(samples_bernoulli)
kurt_binomial = scipy.stats.kurtosis(samples_binomial)
kurt_poisson = scipy.stats.kurtosis(samples_poisson)

print(f'{kurt_bernoulli}')
print(f'{kurt_binomial}')
print(f'{kurt_poisson}')
```

```python
skew_bernoulli = scipy.stats.skew(samples_bernoulli)
skew_binomial = scipy.stats.skew(samples_binomial)
skew_poisson = scipy.stats.skew(samples_poisson)

print(f'{skew_bernoulli}')
print(f'{skew_binomial}')
print(f'{skew_poisson}')
```

### Zadanie 4

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))


ax1.hist(samples_bernoulli, bins=2, density=True, alpha=0.6, label="Histogram")
x1 = np.linspace(0, 1, 10)
y1 = scipy.stats.bernoulli.pmf(x1, p=0.5)
ax1.plot(x1, y1, 'r--', label="Rozkład prawdopodobieństwa")
ax1.set_xlabel("Wartość")
ax1.set_ylabel("Prawdopodobieństwo")
ax1.legend()
ax1.set_title("Rozkład Bernoulliego")


ax2.hist(samples_binomial, density=True, alpha=0.6, label="Histogram")
x2 = np.arange(0, 11)
y2 = scipy.stats.binom.pmf(x2, n=10, p=0.5)
ax2.plot(x2, y2, 'r--', label="Rozkład prawdopodobieństwa")
ax2.set_xlabel("Wartość")
ax2.set_ylabel("Prawdopodobieństwo")
ax2.legend()
ax2.set_title("Rozkład Dwumianowy")


ax3.hist(samples_poisson, bins=2, density=True, alpha=0.6, label="Histogram")
x3 = np.arange(0, 11)
y3 = scipy.stats.poisson.pmf(x3,mu=10)
ax3.plot(x2, y2, 'r--', label="Rozkład prawdopodobieństwa")
ax3.set_xlabel("Wartość")
ax3.set_ylabel("Prawdopodobieństwo")
ax3.legend()
ax3.set_title("Rozkład Poissona")
```
